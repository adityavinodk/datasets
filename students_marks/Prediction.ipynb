```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
import scipy as sc
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df
```

```python
plt.hist(x=df.gender)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

plt.hist(x=df.race)
plt.xlabel('Race')
plt.ylabel('Count')
plt.show()
```

```python
# Math Scores
sns.boxplot(x='gender', y='math_score', data=df)
plt.show()

grid = sns.FacetGrid(df, col="gender")
grid.map(plt.hist, "math_score")
plt.show()
```

```python
plt.scatter(x=df.math_score, y=df.gender)
plt.show()
```
